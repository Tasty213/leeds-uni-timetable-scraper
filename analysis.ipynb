```python
import pandas as pd
data = pd.read_csv("output.csv", names=["Name", "Start", "End", "Info", "Location", "Timetable"])
data.head()
```

```python
data["Start"] = pd.to_datetime(data["Start"], utc=True)
data["End"] = pd.to_datetime(data["End"], utc=True)
data.info()
```

```python
import seaborn as sns
```

```python
data["Day"] = data["Start"].dt.day_name()
data["Hour"] = data["Start"].dt.hour
data["Day Hour"] = data["Day"].astype(str) + " " + data["Hour"].astype(str)
sns.barplot(data, x="Hour")
```

```python
from matplotlib import pyplot as plt

distribution = pd.DataFrame(data["Day Hour"].value_counts())
distribution.rename(columns = {"Day Hour":"Count"}, inplace = True)
distribution["Distribution"] = (distribution["Count"]/2024763)*100
distribution.head()

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))

g = sns.barplot(distribution, x="Count", y=distribution.index)
g.tick_params(axis="x", rotation=90)
```
